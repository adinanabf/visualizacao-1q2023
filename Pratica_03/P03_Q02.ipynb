### Cálculo do percentual acumulado

```python
df['2000-2007'] = (df['2007'] /df['2000']) - 1
df['2007-2014'] = (df['2014']/df['2007']) - 1
df['2014-2022'] = (df['2022']/df['2014']) - 1

print('\
    Taxa = 0: Não houve alteração no GHI no período de tempo analisado\n\
    Taxa > 0: Houve melhora no GHI\n\
    Taxa < 0: Houve piora no GHI')
```

```python
df['2000-2022'] = ((1 + df['2000-2007']) * (1 + df['2007-2014']) * (1 + df['2014-2022'])) - 1
df
```

```python
search_index = df[df['Country'] == "Côte d'Ivoire"].iloc[0].name
df.loc[search_index, 'Country'] = "Ivory Coast"

search_index = df[df['Country'] == "Türkiye"].iloc[0].name
df.loc[search_index, 'Country'] = "Turkey"

search_index = df[df['Country'] == "Lao"].iloc[0].name
df.loc[search_index, 'Country'] = "Laos"

search_index = df[df['Country'] == "Korea"].iloc[0].name
df.loc[search_index, 'Country'] = "North Korea"
```

```python
import pycountry_convert as pcountry

def get_continent_code(row):
    country_code = pcountry.country_name_to_country_alpha2(row.Country, cn_name_format='default')
    if country_code == 'TL':
        return 'Timor-Leste'
    return pcountry.country_alpha2_to_continent_code(country_code)

def get_continent_name(row):
    if row.Continent == 'Timor-Leste':
        return 'Asia'
    return pcountry.convert_continent_code_to_continent_name(row.Continent)
```

```python
df['Continent'] = df.apply(get_continent_code, axis=1)
df['Continent'] = df.apply(get_continent_name, axis=1)
df
```
